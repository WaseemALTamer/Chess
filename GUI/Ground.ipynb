```python
import pygame
pygame.init()

screen = pygame.display.set_mode((400, 400))


running = True
while running:
    for e in pygame.event.get():
        if e.type == pygame.QUIT:
            running = False

    screen.fill((0, 0, 0))

    s = pygame.Surface((200, 200), pygame.SRCALPHA)
    s.fill((255, 0, 0, 50))  # very transparent red
    screen.blit(s.convert_alpha(), (100, 100))

    pygame.display.update()


pygame.quit()
```
